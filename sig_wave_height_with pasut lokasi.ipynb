```python
import matplotlib.pyplot as plt
import mikeio
import mikeio.generic as generic
import numpy as np

dfs = mikeio.open("D:\gel.dfsu")
dfs
ds = dfs.read()
ds

wl_laststep = ds["Sign. Wave Height"].isel(time=-1)
ax = wl_laststep.plot(cmap="viridis", show_mesh=True, figsize=(12,12),  label="Significant Wave Height Spring Tide", levels=np.arange(0, 0.22, 0.025))
ax.scatter(x=123, y=-0.8, marker='*', s=300, label="Location of Tide Measurement", color='red')  
ax.legend();

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize = (17,5))

da = ds["Sign. Wave Height"].isel(time=-4)
da.plot.contourf(ax=ax1, title="Sign. Wave Height Spring Tide", levels=np.arange(0, 0.21, 0.02))
#ax1.scatter(x=123.0938, y=-0.843, marker='*', s=300, label="Location of Tide Measurement") 
#ax1.legend();
da = ds["y-comp. of wave vector"].isel(time=-2)
da.plot.contourf(ax=ax2, title="V velocity Spring Tide", label="Luwuk Waters V velocity [m/s]", levels=np.arange(-0.033, 0.050, 0.01))

##################################################
fig, (ax3, ax4) = plt.subplots(ncols=2, figsize = (17,5))

da = ds["Sign. Wave Height"].isel(time=-4)
da.plot.contourf(ax=ax3, title="Sign. Wave Height Spring Tide",levels=np.arange(0, 0.21, 0.02))

da = ds["x-comp. of wave vector"].isel(time=-2)
da.plot.contourf(ax=ax4, title="U velocity Spring Tide", label="Luwuk Waters U velocity [m/s]",levels=np.arange(-0.25, 0, 0.02));

##########################################


##########################################
ds = mikeio.read("D:\gel.dfsu")
ds

dsm  = ds.mean()
mean_ds = dsm["Sign. Wave Height"]
mean_ds.shape

mean_ds.plot(title="Mean Sign. Wave Height",levels=np.arange(0, 0.21, 0.025));


df = ds.mean(axis="space").to_dataframe()
df.head()

ds["Sign. Wave Height"].max(axis="space").to_dataframe().plot(title="Max Sign. Wave Height");

#########################################


```
